### Reading images using opencv
- **retval = cv2.imread(filename,[flags])**
Flags
- cv2.IMREAD_GRAYSCALE or 0 : loads image in grayscale mode
- cv2.IMREAD_COLOR or 1 : loads a color image, default flag
- cv2.IMREAD_UNCHANGED or -1 : loads image as such includinf alpha  channel

```python
# Read image as grayscale
gs_image = cv2.imread('checkerboard_18x18.png',0)
# Print the image data (values of the 2d numpy array)
print(gs_image) # Each pixel value(8 bits) is between 0 to 255
```

### Splitting and Merging color channels
- **cv2.split()** : split multichannel array into single channel array
- **cv2.merge()** : merge single channel array into multichannel array

```python
# Reverse the original image before any further transformation
nature_reversed = cv2.imread('Nature.jpeg',1)
nature_bgr = nature_reversed[:,:,::-1]
# Split the components of image into R,G,B
b,g,r = cv2.split(nature_bgr)

# Merged image equal to the original image
nature_merged = cv2.merge((b,g,r))
```

```python
# Show the channels
plt.figure(figsize=[20,6])
plt.subplot(231); plt.imshow(r,cmap='gray');plt.title('Red Channel');
plt.subplot(232); plt.imshow(g,cmap='gray');plt.title('Green Channel');
plt.subplot(233); plt.imshow(b,cmap='gray');plt.title('Blue Channel');
plt.subplot(234); plt.imshow(nature_bgr);plt.title('Original');
plt.subplot(235); plt.imshow(nature_reversed);plt.title('Reversed');
plt.subplot(236); plt.imshow(nature_merged);plt.title('Merged');
```

### Converting to different color spaces
- Converts an image from one color space to another.

**dst = cv2.cvt.Color(src,code)**
- src =  source image
- code = color space conversion code (ColorConversionCodes)

###  Changing form BGR  to RGB

```python
# Read the image
n_bgr = cv2.imread('Nature.jpeg')
# Convert the image to RGB
n_rgb = cv2.cvtColor(n_bgr,cv2.COLOR_BGR2RGB)
# Show the image
plt.figure(figsize=[15,6])
plt.subplot(121); plt.imshow(n_bgr);plt.title('BGR image');
plt.subplot(122); plt.imshow(n_rgb);plt.title('RGB image');
```

### Changing to HSV color space

```python
# Convert BGR to HSV
n_hsv = cv2.cvtColor(n_bgr,cv2.COLOR_BGR2HSV)
# Split into different components of HSV
h,s,v = cv2.split(n_hsv)

plt.figure(figsize=[15,6])
plt.subplot(231); plt.imshow(n_bgr);plt.title('BGR image');
plt.subplot(232); plt.imshow(n_hsv);plt.title('HSV image');

plt.subplot(233); plt.imshow(h,cmap='gray');plt.title('Hue image'); # Color of the image
plt.subplot(234); plt.imshow(s,cmap='gray');plt.title('Saturation image'); # Intensity of the image
plt.subplot(235); plt.imshow(v,cmap='gray');plt.title('Value image');# Value of the image
```

### Mdifying individual channel

```python
# Incrment the value of hue
h_new = h+100 
# Merge with saturation and value channel
n_hsv_merged = cv2.merge((h_new,s,v)) 
# Convert it to RGB
n_hsv_merged_rgb = cv2.cvtColor(n_hsv_merged,cv2.COLOR_HSV2RGB) 
```

```python
# Show the new channels
plt.figure(figsize=[15,6])
plt.subplot(231); plt.imshow(n_hsv_merged);plt.title('New HSV image');
plt.subplot(232); plt.imshow(n_hsv);plt.title('HSV image');

plt.subplot(233); plt.imshow(h,cmap='gray');plt.title('Hue image'); # Color of the image
plt.subplot(234); plt.imshow(h_new,cmap='gray');plt.title('New_Hue image');
plt.subplot(235); plt.imshow(n_hsv_merged_rgb);plt.title('New RGB  image');
plt.subplot(236); plt.imshow(n_rgb);plt.title('Original image');
```
