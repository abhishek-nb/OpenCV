# Image Manipulation
- Accessing and Manipulating image pixels
- Resizing
- Cropping
- Mapping

```python
# Importing libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from IPython.display import Image
```

```python
# Image read and load
cb_img = cv2.imread('checkerboard_18x18.png',0)
plt.imshow(cb_img,cmap='gray')
print(cb_img)
```

### Accessing individual pixels
- For accessing any pixel location write such as matrix notaion [r,c]. Where 'r' is row number and 'c' is column number of matrix.

```python
# Pixel value of the first black box
print(cb_img[0,0])
# Pixel value of the first white box
print(cb_img[0,7])
```

```python
# Modifying image pixels
cb_img_copy = cb_img.copy()
cb_img_copy[2,0] = 255  
cb_img_copy[2,1] = 255
cb_img_copy[2,2] = 255
cb_img_copy[2,3] = 255
cb_img_copy[2,4] = 255
cb_img_copy[2,5] = 255

cb_img_copy[16:4,16:4] = 255 

plt.imshow(cb_img_copy,cmap='gray')
```

```python

cb_img_copy[8,9] = 150
cb_img_copy[8,10] = 150

plt.imshow(cb_img_copy,cmap='gray')
```

### Cropping of the image

```python
nz_bgr = cv2.imread('New_Zealand_Boat.jpg',1)
nz_rgb = cv2.cvtColor(nz_bgr,cv2.COLOR_BGR2RGB)
plt.imshow(nz_rgb)
```

```python
c_nz = nz_rgb[200:400,300:600]
plt.imshow(c_nz)
```

### Resizing

**resize(src, dsize[, dst[, fx[, fy[, interpolation]]]])** 
- src = input image size
- dsize = output image size
- fx = scaling in horizontal direction
- fy = scaling in vertical direction

### Method 1
- Specifying scaling factor using fx and fy

```python
resized_cropped_region_2x = cv2.resize(c_nz,None,fx = 2,fy = 2)
plt.imshow(resized_cropped_region_2x)
```

### Method 2 :
- Specifying exact size of the output image

```python
d_width = 100
d_height = 200
dim = (d_width,d_height)
```

```python
resized_cropped_region = cv2.resize(c_nz,dsize = dim,interpolation = cv2.INTER_AREA)
plt.imshow(resized_cropped_region)
```


- Resize while maintaing aspect ratio

```python
# Aspect ratio
d_width = 100
aspect_ratio = d_width / c_nz.shape[1]
d_height = int(c_nz.shape[0]*aspect_ratio)
dim = (d_width,d_height)
```

```python
resized_cropped_region=cv2.resize(c_nz,dsize = dim, interpolation= cv2.INTER_AREA)
plt.imshow(resized_cropped_region)
```

```python
# Swap channel order 
resized_cropped_region_2x = resized_cropped_region_2x[:,:,::-1]
# Savee rsized image to disk
cv2.imwrite("resized_cropped_region_2x.png",resized_cropped_region_2x)
```

```python
plt.imshow(cv2.imread('resized_cropped_region_2x.png'))
```

### Flipping
**dst = cv2.flip(src,code)**
- src = input image
- code = 0 means flipping around the x-axis,
- Positive value (for example, 1) means flipping around y-axis. 
- Negative value (for example, -1) means flipping
around both axes.

```python
nz_rgb_h_f = cv2.flip(nz_rgb,1) # Horizontal flip
nz_rgb_v_f = cv2.flip(nz_rgb,0) # Vertical flip
nz_rgb_hv_f = cv2.flip(nz_rgb,-1) # Both flip
```

```python
plt.figure(figsize=[10,6])
plt.subplot(231); plt.imshow(nz_rgb_h_f);plt.title('Horizontal Flip');
plt.subplot(232); plt.imshow(nz_rgb_v_f);plt.title('Vertical Flip');
plt.subplot(233); plt.imshow(nz_rgb_hv_f);plt.title('Both Flipped');
plt.subplot(234); plt.imshow(nz_rgb);plt.title('Original');
```
